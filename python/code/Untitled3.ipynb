```python
%matplotlib inline

import madmom
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage.filters import maximum_filter
import librosa
import IPython.display as ipd 
```

```python
y, sr = librosa.load('madmom/153036.wav')
print('sr', sr)
```

```python
o_env = librosa.onset.onset_strength(y, sr=sr)
```

```python
onset_frames = librosa.onset.onset_detect(onset_envelope=o_env, sr=sr)
```

```python
print(onset_frames)
```

```python
onset_frames_times = librosa.frames_to_time(onset_frames, sr=sr)
print(onset_frames_times)
```

```python
peaks = librosa.util.peak_pick(o_env, 3, 3, 3, 5, 0.2, 10)
```

```python
print(peaks)
```

```python
peak_times = librosa.frames_to_time(peaks, sr=sr)
print(peak_times)
```
